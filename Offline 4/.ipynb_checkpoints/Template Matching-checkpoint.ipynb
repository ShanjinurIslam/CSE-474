```python
import cv2
```

```python
vidObj = cv2.VideoCapture('movie.mov') 

frames = []
count = 0
success = 1

while success: 
    success, image = vidObj.read()     
    if image is None:
        break
    
    
    frames.append(cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)/255)
```

```python
reference = cv2.imread('reference.jpg')

reference = cv2.cvtColor(reference, cv2.COLOR_BGR2GRAY)/255

frames[0].shape
```

```python
def ExhaustiveSearch(t,r):
    M,N = r.shape[0],r.shape[1]
    I,J = t.shape[0],t.shape[1]
    
    count = 0
    
    best_node = [0,0]
    max_correlation = 0
    
    for m in range(I-M+1):
        for n in range(J-N+1):
            print(m,n)
            current_correlation = 0
            for i in range(m,m+M-1):
                for j in range(n,n+N-1):
                    current_correlation += t[i][j]*r[i-m][j-n]
            
            if(current_correlation>max_correlation):
                max_correlation = current_correlation
                best_node = [m,n]
    
    print(best_node)
```

```python
ExhaustiveSearch(frames[0],reference)
```
