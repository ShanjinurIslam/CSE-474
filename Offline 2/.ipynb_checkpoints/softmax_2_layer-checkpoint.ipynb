```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('trainNN.txt')

lines = file.readlines()

X = []
labels = []


for line in lines:
    var = [float(x) for x in line.split()]
    
    X.append(var[:-1])
    labels.append(int(var[-1]))
```

```python
number_of_classes = max(labels)
N = len(X)

Y = np.zeros((N,number_of_classes))

for i in range(N):
    Y[i,labels[i]-1] = 1
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def sigmoid_derivative(z):
    return sigmoid(z) *(1-sigmoid (z))

def softmax(z):
    e = np.exp(z)
    return e / e.sum(axis=1, keepdims=True)
```

```python
input_vector = X.shape[1]
output_vector = Y.shape[1]

print(input_vector,output_vector)
```

```python
hidden_nodes = 4

# hidden layer

wh = np.random.uniform(-1,1,(input_vector,hidden_nodes))
bh = np.zeros(hidden_nodes)

# output layer

wo = np.random.uniform(-1,1,(hidden_nodes,output_vector))
bo = np.zeros(output_vector)

learning_rate = 10e-4

error_cost = []

bo
```

```python
for i in range(500):
    zh = np.dot(X, wh) + bh
    ah = sigmoid(zh)

        # Phase 2
    zo = np.dot(ah, wo) + bo
    ao = softmax(zo)

    dcost_dzo = ao-Y
    dzo_dwo = ah

    dcost_wo = np.dot(dzo_dwo.T, dcost_dzo)
    dcost_bo = dcost_dzo


    dzo_dah = wo
    dcost_dah = np.dot(dcost_dzo , dzo_dah.T)
    dah_dzh = sigmoid_derivative(zh)
    dzh_dwh = X
    dcost_wh = np.dot(dzh_dwh.T, dah_dzh * dcost_dah)

    dcost_bh = dcost_dah * dah_dzh
    

    wh -= learning_rate * dcost_wh
    bh -= learning_rate * dcost_bh.sum(axis=0)

    wo -= learning_rate * dcost_wo
    bo -= learning_rate * dcost_bo.sum(axis=0)
    
    print(bh)
```
